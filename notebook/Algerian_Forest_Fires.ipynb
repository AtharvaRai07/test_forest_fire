#Model Traning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/Algerian_Dataset_Cleaned.csv')
```

```python
df.head()
```

```python
df.drop(['day','month','year'],axis = 1,inplace = True)
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
X = df.drop('FWI',axis = 1 )
y = df['FWI']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr, annot = True)
```

```python
def correlations (dataset, threshold):
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) > threshold:
        colname = corr_matrix.columns[i]
        col_corr.add(colname)
  return col_corr
```

```python
corr_feautures = correlations(X_train,0.85)
```

```python
X_train.drop(corr_feautures, axis = 1, inplace = True)
X_test.drop(corr_feautures, axis = 1, inplace = True)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_Scaled  = sc.fit_transform(X_train)
X_test_Scaled = sc.transform(X_test)
```

```python
plt.subplots(figsize=(13,6))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_ train Before Scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_Scaled)
plt.title("X_ train After Scaling")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train_Scaled,y_train)
```

```python
y_pred = lr.predict(X_test_Scaled)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
r2_score(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```

#Lasso Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train_Scaled,y_train)
```

```python
y_pred = lasso.predict(X_test_Scaled)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
r2_score(y_test,y_pred)
```

```python
lassocv.alpha_
```

```python
y_pred = lassocv.predict(X_test_Scaled)
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

#Ridge Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_Scaled,y_train)

y_pred = ridge.predict(X_test_Scaled)

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))

plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
```

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv = 5)
ridgecv.fit(X_train_Scaled,y_train)

y_pred = lassocv.predict(X_test_Scaled)
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print('MAE:',mean_absolute_error(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
